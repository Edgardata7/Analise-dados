## Análise Univariada (uma variável)

```python
# 2. Separe manualmente as variáveis numéricas
colunas_numericas = ['age', 'trtbps', 'chol', 'thalachh', 'oldpeak', 'caa']

for coluna in colunas_numericas:
    plt.figure(figsize=(8, 6))  # Tamanho da figura do gráfico
    plt.hist(df[coluna], bins=30, edgecolor='k')  # O número de bins pode ser ajustado conforme sua preferência
    plt.xlabel(coluna)  # Nomeie o eixo x com o nome da coluna
    plt.ylabel('Frequência')  # Nomeie o eixo y
    plt.title(f'Histograma de {coluna}')  # Adicione um título ao gráfico
    plt.grid(True)  # Adicione linhas de grade
    plt.show()  # Exibir o gráfico
```

```python
# variaveis categóticas
variaveis_categoricas = ['sex', 'cp', 'fbs', 'restecg', 'exng', 'slp', 'caa', 'thalachh']

# plotar gráficos de frequencia para as variaveis categóricas

for var in variaveis_categoricas:
  plt.figure(figsize=(7, 5))
  df[var].value_counts() .sort_index() .plot(kind='bar', rot=0)
  plt.title(f'Frequência de Valores para {var}')
  plt.xlabel(var)
  plt.ylabel('Frequência')
  plt.show()
```

## Análise Bivariada (duas ou mais variáveis)

```python
# Variáveis categóticas
variaveis_categoricas = ['sex', 'cp', 'fbs', 'restecg', 'exng', 'slp', 'caa', 'thall',]


# plotar gráfico de frequência para as variáveis categóricas

for var in variaveis_categoricas:
  plt.figure(figsize=(7, 8))
  df.groupby([var,'output']).size().unstack().plot(kind='bar',stacked=False)
  plt.title(f'Frequência de {var} por output')
  plt.xlabel(var)
  plt.ylabel('frequência')
  plt.legend(['No Disease', 'Disease'])
  plt.show()
```

```python
# Calcular a matriz de correlação
corr_matrix = df.corr() # substitua "df" pelo nome do seu dataframe

# Criar o Heatmap de correlação
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm") # configura o heatmap

# Exibir o plot
plt.title("Heatmap de Correlação") # define o titulo do gráfico
plt.figure(figsize=(10, 8)) # define o tamanha da figura
plt.show() # exibe o gráfico
```
